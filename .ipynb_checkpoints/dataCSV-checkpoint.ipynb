```python
import FeatureExtraction as fe
def getQM(t):
    return t.questionStat

def getKey(t, domain, key):
    return t.keywordStat[domain][key]    
```

```python
import pandas as pd
domainIndexOfKeywords = {  "Agenda":["agenda","you want to","add anything to the agenda"] \
                         , "Feedback":["learn","goal"] \
                         , "Homework":["homework","homework review","in the way","assign"] \
                         , "Focus on Key Cognitions":["think","thought","make you","you do"]  }           
index = ["T1","T3","T4","T5","T6","T7"]
columns = ["tech score","question_mark","agenda","learn","goal","homework","homework review","in the way" \
           ,"assign","think","thought","make you","you do","you want to" \
           ,"add anything to the agenda"]
  
fileName = ["Transcript1.txt","Transcript3.txt","Transcript4.txt","Transcript5.txt","Transcript6.txt","Transcript7.txt"]
    
df = pd.DataFrame(index = index, columns = columns)
with open('numericAttributes.csv', 'a') as f:
    df.to_csv(f, header = False)

df["tech score"] = [44.0,21.0,38.0,29.0,28.0,11.0]
for n in range(len(fileName)):
    t1 = fe.KeywordCount(fileName[n])
    t1.updateWordCount()
    t1.updateKeywordStatistics()
    t1.questionMark()
    t1.emoVoice()
    t1.keywordQuestionmark()
    t1.nonSpeech()
    t1.lengthByTurn()
    t2 = fe.LabelledKeywordCount(fileName[n],'p')
    t2.updateWordCount()
    t2.updateKeywordStatistics()
    t2.questionMark()
    t2.emoVoice()
    t2.keywordQuestionmark()
    t2.nonSpeech()
    t2.lengthByTurn()
    df["question_mark"][n] = float(getQM(t1))
    for domain in domainIndexOfKeywords.keys():
        for kw in domainIndexOfKeywords[domain]:
            df[kw][n] = float(getKey(t2, domain, kw))

df = df.convert_objects(convert_numeric=True)
df
```

```python
corr_matrix = df.corr()
corr = corr_matrix["tech score"].sort_values(ascending=False)
corr.to_csv("StdCorrCoeff.csv")
writer = pd.ExcelWriter('StdCorrCoeff.xlsx')
corr.to_excel(writer,'Sheet1')
writer.save()
```
